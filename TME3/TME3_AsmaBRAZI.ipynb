#  <center>Algorithme de descente de gradient</center>

### Coder une fonction optimize(fonc,dfonc,xinit,eps,max_iter)

```python
import numpy as np
import pandas as pd
from numpy import random
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
def optimize(fonc,dfonc,xinit,eps,max_iter):
    x_histo,f_histo,grad_histo=[],[],[]
    x_histo.append(xinit)
    x=xinit
    for i in range(max_iter):
        f_histo.append(fonc(x))
        df_x= dfonc(x)
        grad_histo.append(df_x)
        x=x-eps*df_x
        x_histo.append(x)
    #the last value of f(x) and df(x)
    f_histo.append(fonc(x))
    df_x= dfonc(x)
    grad_histo.append(df_x)
    return np.array(x_histo),np.array(f_histo),np.array(grad_histo)
```

### DÃ©finition de fonctions

```python
def fonc1(x):
    return x*np.cos(x)

def dfonc1(x):
    return np.cos(x)-x*np.sin(x)
```

```python
def fonc2(x):
    return -np.log(x)+x**2

def dfonc2(x):
    return -1/x+2*x
```

```python
def fonc3(x):
    x1,x2 =x[0], x[1]
    return 100*(x2-x1**2)**2+(1-x1)**2

def dfonc3(x):
    x1,x2 =x[0], x[1]
    return np.array([-400*x1*(x2-x1**2)+(1-x1)**2,-199*x1**2-2*x1+200*x2+1]).T    
```

```python
def filter_data(datax, datay, a,b):
    i = np.where(np.logical_or(datay == a, datay == b))[0]
    new_datay= np.array(datay[i])
    new_datay[new_datay == a] = -1
    new_datay[new_datay == b] = 1
    return np.array(datax[i]),new_datay

def divideToBins(datax, bins, res=0.0001):
        min, max = datax.min(), datax.max()
        return ((datax - min)/(max - min)*bins - res).astype(int)
```

```python
class Logistic_regression:
    def __init__(self, max_iter, eps):
        self.max_iter = max_iter
        self.eps = eps
        
    def fit(self, datax, datay):
        self.w = np.zeros(datax.shape[1])
        self.w = self.optimize(self.eps,self.max_iter, self.w,datax,datay)
        
    def sigmoid(self,x):
        sig= 1.0/(1.0 + np.exp(-np.dot(self.w.T,x)))
        return 1 if sig < 0.5 else -1

    def predict(self, datax):
        return np.sign(np.dot(self.w, datax.T))
    
    def score(self, datax, datay):
        return np.mean(self.predict(datax) == datay)
    
    def fonc(self, w,datax,datay):
        result = 0.0
        for i in range(len(datax)):
            result += np.log(1 + np.exp(-(2*datay[i]-1) * np.dot(w.T,datax[i])))
        return result/len(datax)
    
    def dfonc(self, w,datax,datay):
        r = 0.0
        for i in range(len(datax)):
            r += (-(2*datay[i]-1) * datax[i] * np.exp(-(2*datay[i]-1) * np.dot(w.T,datax[i])))/(1 + np.exp(-(2*datay[i]-1) * np.dot(w.T,datax[i])))
        return r/len(datax)
           
    def optimize(self,eps,max_iter,w,datax,datay):
        ww = w.copy()
        for i in range(max_iter):
            df_w=self.dfonc(ww,datax,datay)
            ww=ww-eps*df_w
        return ww
```

```python
class NaiveBayes():
    def __init__(self, bins=4):
        self.bins = bins
        
    def fit(self, datax, datay):
        self.X = datax
        self.Y = datay
        Ny=len(datay)
        # p(Y)
        y = np.unique(datay, return_counts=True)
        self.Py = y[1]/Ny
        datax = divideToBins(datax, self.bins)
        Pxy0 = np.unique(datax[datay == -1], return_counts=True)[1]/y[1][0]
        Pxy1 = np.unique(datax[datay == 1], return_counts=True)[1]/y[1][1]
        self.Pxy = np.vstack((Pxy0, Pxy1))
    
    def predict(self, datax):
        datax = divideToBins(datax, self.bins)
        res = np.zeros(len(datax))
        for i in range(len(datax)):
            res[i] = 2* np.argmax([np.sum([np.log(self.Pxy[0][x]) for x in datax[i]]),  np.sum([np.log(self.Pxy[1][x]) for x in datax[i]])]) - 1
        return res
    
    def score(self, datax, datay):
        return np.mean(self.predict(datax) == datay) 
    
    
    
```
